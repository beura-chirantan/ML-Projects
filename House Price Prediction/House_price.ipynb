#IMPORTING LIBRARIES

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

#IMPORTING DATASET

```python
dataset=pd.read_csv('california_housing.csv')
x=dataset.drop(['median_house_value'], axis=1)  # axis=1 indicates that we are dropping one column
y=dataset['median_house_value']
```

#ANALYZING THE DATASET

```python
dataset
```

```python
dataset.info()
```

#DROPPING NAN ENTRIES

```python
dataset.dropna(inplace=True)  #drops the nan entris and saves the results in the data object
```

```python
dataset.info()
```

#SPLITTING DATASET INTO TRAINING AND TESTING SET

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=42)
```

#ANALYZING THE TRAINING SET

```python
train_dataset=x_train.join(y_train)  # join method is used to combine 2 dataframes
```

```python
train_dataset
```

#VISUALISING THE TRAINING SET

```python
train_dataset.hist(figsize=(15,8))  # to get histograms for the different features
```

```python
train_dataset.corr()  #gives a correlation matrix
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_dataset.corr(), annot=True, cmap="YlGnBu")  # annotation=true so that we can see the correlation numbers. Color map=YellowGreenBlue

# from the map we can see that median_house_value relate quite strongly with median_house_value so it is great predictor for house value
```

```python
# datas in total_rooms, total_bedrooms, population and households are skewed and not proper gaussian bell curves in the histogram. So we will take logarithm of those features

train_dataset['total_rooms']=np.log(train_dataset['total_rooms'] + 1)  # inorder to prevent 0 values we add 1
train_dataset['total_bedrooms']=np.log(train_dataset['total_bedrooms'] + 1)
train_dataset['population']=np.log(train_dataset['population'] + 1)
train_dataset['households']=np.log(train_dataset['households'] + 1)
```

```python
train_dataset.hist(figsize=(15,8))
```

#ENCODING THE OCEAN_PROXIMITY

```python
train_dataset.ocean_proximity.value_counts()
```

```python
train_dataset=train_dataset.join(pd.get_dummies(train_dataset.ocean_proximity)).drop(['ocean_proximity'], axis=1)
# we will encode ocean_proximity using OneHotEncoding using pandas and will create a new feature for all these categories
# then we will drop ocean_proximity column and save the results in train_dataset
```

```python
train_dataset
```

```python
# finding correlation between features ocean_proximity and meadin_house_value
plt.figure(figsize=(15,8))
sns.heatmap(train_dataset.corr(), annot=True, cmap="YlGnBu")
```

```python
train_dataset['bedroom_ratio']=train_dataset['total_bedrooms'] / train_dataset['total_rooms']
train_dataset['household_rooms']=train_dataset['total_rooms'] / train_dataset['households']
# we are creating more featues like this in order to have a better grip on our model and make our model better
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_dataset.corr(), annot=True, cmap="YlGnBu")
```

#TRAINING AND FEATURE SCALING OUR MODEL USING RANODM FOREST REGRESSION

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()

x_train, y_train=train_dataset.drop(['median_house_value'], axis=1), train_dataset['median_house_value']
x_train_sc=sc.fit_transform(x_train)

regressor=RandomForestRegressor(random_state=42)
regressor.fit(x_train_sc, y_train)
```

```python
test_dataset=x_test.join(y_test)

test_dataset['total_rooms']=np.log(test_dataset['total_rooms'] + 1)
test_dataset['total_bedrooms']=np.log(test_dataset['total_bedrooms'] + 1)
test_dataset['population']=np.log(test_dataset['population'] + 1)
test_dataset['households']=np.log(test_dataset['households'] + 1)

test_dataset=test_dataset.join(pd.get_dummies(test_dataset.ocean_proximity)).drop(['ocean_proximity'], axis=1)

test_dataset['bedroom_ratio']=test_dataset['total_bedrooms'] / test_dataset['total_rooms']
test_dataset['household_rooms']=test_dataset['total_rooms'] / test_dataset['households']
```

#DROPPING NAN VALUES FROM TESTSET

```python
test_dataset.dropna(inplace=True)
```

```python
x_test, y_test=test_dataset.drop(['median_house_value'], axis=1), test_dataset['median_house_value']
```

```python
x_test_sc=sc.transform(x_test)
```

#GETTING ACCURACY SCORE

```python
accuracy=regressor.score(x_test_sc, y_test)
print(f"Accuracy: {accuracy:.2f}")
```
